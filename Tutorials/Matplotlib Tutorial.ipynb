```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### using this command to show plots in the notebook

```python
%matplotlib inline
```

## let's import the data and start visualizing it

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset
```

### let's start by drawing just the salaries vs the level

```python
salaries = dataset.iloc[:,-1].values
levels = dataset.iloc[:, -2].values
plt.plot(salaries,levels)
plt.show()
```

### What about the opposite ?

```python
plt.plot(levels, salaries)
plt.show()
```

#### we can easly see that the plot takes the x-axis values as the first argument then the y-axis by looking at the above example 

## We can change the shape of the relation !

```python
plt.plot(levels, salaries, 'ro')
plt.show()
```

```python
plt.plot(levels, salaries, linewidth = 7.0)
plt.show()
```

## Naming your axis is so important dude !

```python
plt.plot(levels, salaries)
plt.xlabel('level')
plt.ylabel('salary')
plt.title('levels vs salary !')
plt.show()
```

## Now let's make a new different salaries and compare them with our salaries

```python
new_salaries = salaries + (np.random.rand(np.shape(salaries)[0]))*90000
new_salaries
```

### Let's take a look to our new data 

```python
plt.plot(levels, new_salaries)
plt.show()
```

### Now let's compare them together

```python
plt.plot(levels, salaries, levels, new_salaries)
plt.show()
```

### Notice that the new salaries are always above the salaries as we added a random values to it
