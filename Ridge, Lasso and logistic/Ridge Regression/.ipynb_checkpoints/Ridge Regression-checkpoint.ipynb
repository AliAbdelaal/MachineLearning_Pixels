```python
# import libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Salary_Data.csv')
features =  dataset.iloc[:,:-1].values
salarys = dataset.iloc[:,-1].values
dataset
```

## Now let's generate the same poly features as before !

```python
# define a function that return whatever degree polynomial you want
def getPolynomial(data,degree=1):
    if degree == 1 :
        return data 
    from sklearn.preprocessing import PolynomialFeatures
    poly = PolynomialFeatures(degree= degree)
    poly_features = poly.fit_transform(data)
    return poly_features
# define a function for plotting
def plot(feature = [], actual = [], predicted = [], alpha = 1):
    plt.plot(feature, actual, '.', feature, predicted, '-')
    plt.title('alpha = {}'.format(alpha))
    plt.show()
```

```python
poly_features = getPolynomial(features, 13)
pd.DataFrame(poly_features)
```

# Now let's go Ridge !
## Remember that ridge change the cost function and it now depends not only on the error metric(RSS for example) but it also depends on the weights of the coeff.
### cost = RSS + alpha x (L2)
where l2 is the sum squars of the coeff(s)

```python
coeff_matrix = []
for alpha in [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]:
    from sklearn.linear_model import Ridge
    ridge = Ridge(alpha = alpha, normalize=True)
    ridge.fit(poly_features, salarys)
    coeff_matrix.append(ridge.coef_)
    plot(features, salarys, ridge.predict(poly_features), alpha=alpha)
    print('------------------')
pd.DataFrame(coeff_matrix)
```

## We can see clearly how ridge regression minimized the weights of the features

### Notice also how the weights didn't went to zero!
