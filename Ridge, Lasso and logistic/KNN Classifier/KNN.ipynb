```python
# libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
features = dataset.iloc[:,1:-1].values
goal = dataset.iloc[:,-1].values
dataset
```

```python
# solve the categorical data
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
features[:,0] = encoder.fit_transform(features[:,0])

# standrize the values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features = scaler.fit_transform(features)

# split into train and test
from sklearn.cross_validation import train_test_split
train_set, test_set, goal_train, goal_test = train_test_split(features,goal,train_size =0.8,random_state=0)
```

## Time for the model !

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(train_set, goal_train)
```

## We'll use the Confusion Matrix again here
## till you come up with a better way to represent it ;)

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(goal_test, classifier.predict(test_set))
pd.DataFrame(cm)
```

## But what if we increased K ?

```python
for K in [1, 3, 5, 10, 20, 100, 150, len(train_set)]:
    classifier = KNeighborsClassifier(n_neighbors=K)
    classifier.fit(train_set, goal_train)
    cm = confusion_matrix(goal_test, classifier.predict(test_set))
    print('K = %d\n-------------'%K)
    print(pd.DataFrame(cm))
    print('-------------\n\n')
```
